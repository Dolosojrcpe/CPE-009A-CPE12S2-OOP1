```python

def termCrit(actNumber):
  rScore = []
  tScore = []

  iScore = float(input("Score: "))
  iTotal = float(input("Total: "))

  if iScore > iTotal:
      print("\nInvalid Scores.\n")
      iScore, iTotal = 0,0

  else:
      rScore.append(iScore)
      tScore.append(iTotal)

  grade = sum(rScore) / sum(tScore)
  grade = grade * 100
  return grade
```

```python
# Create a function for the class standing per term
def termCS():

  print("Quizzes:")
  termQuiz = termCrit(int(input("How many quizzes this term?")))

  print("Assignment: ")
  termAss = termCrit(int(input("How many assignments?")))

  grade = (0.3 * termQuiz) + (0.2 * termAss)
  return grade
```

```python
 #Create a function to compute the term grade
def termGrade():
  print("Computing for the term grade.")
  examRaw = float(input("Exam Score: "))
  examTotal = float(input("Exam Total: "))
  examGrade = (examRaw/examTotal) * 100

  CSgrade = termCS()

  grade = (0.5 * examGrade) + (0.5 * CSgrade)
  return grade
```

```python
# Create a function to compute for the grades in the semester
def totalGrade():
  print("Prelim Grade")
  prelimGrade = termGrade()

  print("Midterm Grade")
  midtermGrade = termGrade()
  finalMG = (1/3)*prelimGrade + (2/3)*midtermGrade

  print("Final Grade")
  finalGrade = termGrade()
  finalFG = (1/3)*finalMG + (2/3)*finalGrade

  return finalFG
```

```python
def convertGrade(finalGrade):
  if finalGrade >= 90.0: return "A+"
  elif  finalGrade >= 80.0: return "A"
  elif finalGrade >= 70.0: return "B+"
  elif finalGrade >= 60.0: return "B"
  elif finalGrade >= 50.0: return "C"
  elif finalGrade >= 40.0: return "D"
  elif finalGrade <= 40.0: return "F"




```

```python

print("Overall Results:",convertGrade(totalGrade()))
if totalGrade <= 40.0:
  print("Failed")
elif  totalGrade > 40.0:
  print("Passed")
```

```python
def remainder(n, m):
    while True:
        if n - m < 0:
            # If subtracting m results in a negative number,
            # it means the current 'n' is the remainder.
            return n
        else:
            # Subtract m from n and update n
            n = n - m

# Calculation:
# 1st iteration: 10 - 4 = 6 (Proceed to else)
# 2nd iteration: 6 - 4 = 2 (Proceed to else)
# 3rd iteration: 2 - 4 = -2 (n - m < 0 is True, return n)
print(remainder(10, 4))
```
